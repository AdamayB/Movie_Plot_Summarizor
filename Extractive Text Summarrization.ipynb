# PreProcessing
## Removing special characters
## Removing StopWords

```python
def preProcess(text):
    text=nltk.word_tokenize(text)
    
    #Remove spl chars
    y=[]
    for i in text:
        if i.isalnum():
            y.append(i)
    
    text=y
    
    y=[]
    # remove stopwords : https://www.tutorialspoint.com/Removing-stop-words-with-NLTK-in-Python
    y = [word for word in text if word not in stopwords.words('english')]
    
    return " ".join(y)
```

```python
text=preProcess(text)
```

```python
text
```

```python
words=nltk.word_tokenize(text)
```

## Finding the word that is repeated the most

```python
nlp_words=nltk.FreqDist(words)
```

## Finding the weighted frequency of words
## Weighted freq=word freq/maximum freq

```python
maximum_frequncy = max(nlp_words.values())

for word in nlp_words.keys():
    nlp_words[word] = (nlp_words[word]/maximum_frequncy)
```

```python
nlp_words
```

# <font color=red>Very Important step</font>
## Finding sentence scores

```python
sentence_scores = {}
for sent in sente:
    for word in nltk.word_tokenize(sent.lower()):
        if word in nlp_words.keys():
            if len(sent.split(' ')) < 30:
                if sent not in sentence_scores.keys():
                    sentence_scores[sent] = nlp_words[word]
                else:
                    sentence_scores[sent] += nlp_words[word]
```

```python
print(sorted(sentence_scores.items(),key=lambda x:x[1]))
```

```python
import heapq
```

```python
summary_sentences = heapq.nlargest(7, sentence_scores, key=sentence_scores.get)

summary = ' '.join(summary_sentences)
print(summary)
```
